## LSTM Data Preparation and Feature Engineering

```python
## resampling of data over hour
df_resample = df.resample('h').mean() 
df_resample.shape
```

```python
def prepare_data(data, n_in, n_out, step_foreward):
    df = pd.DataFrame(data)
    
    # check steps and cut data at the beginning
    timesteps = df.shape[0]
    steps = int((timesteps - n_in - n_out) / step_foreward)
    nDrop = timesteps - (steps * n_out + n_in + n_out)
    if nDrop > 0:
        df = df[nDrop:]
    return df
    
def create_tensors(data, column_value, n_in = 1, n_out = 1, step_foreward = 1, dropNaN=True):
    n_features = data.shape[1]
    df = pd.DataFrame(data)

    # X tensor
    cols = list()
    dfX = df[column_value][:-n_out]
    for i in range(n_in, 0, -1):
        cols.append(data.shift(i))
    lagged_df = pd.concat(cols, axis=1)
    if dropNaN:
        lagged_df.dropna(inplace=True)
    # reshape data to shape [samples, timesteps, features]
    tensor_X = lagged_df.values.reshape((lagged_df.shape[0], n_in, n_features))
    tensor_X = tensor_X[range(0, tensor_X.shape[0], step_foreward)]

    # y tensor
    cols = list()
    dfY = df[column_value][n_in:]
    for i in range(0, n_out):
        cols.append(dfY.shift(-i))
    lagged_df = pd.concat(cols, axis=1)
    if dropNaN:
        lagged_df.dropna(inplace=True)
    # reshape data to shape [samples, timesteps]
    tensor_y = lagged_df.values.reshape((lagged_df.shape[0], n_out))
    tensor_y = tensor_y[range(0, tensor_y.shape[0], step_foreward)]

    return tensor_X, tensor_y
```

```python
n_train_time = 365*24
n_in = 24
n_out = 24
step_foreward = n_out

col_names = df_resample.columns
values = df_resample.values[:n_train_time]
scaler = MinMaxScaler(feature_range=(0, 1))
scaled = scaler.fit_transform(values)
df_prep = prepare_data(scaled, n_in, n_out, step_foreward)
df_prep.columns = col_names
train_X, train_y = create_tensors(df_prep, "Global_active_power", n_in, n_out, step_foreward)
```

```python
print(train_X.shape, train_y.shape)
```

```python
train_y[0,:]
```

```python
train_X[1,:,0]
```

```python
model = Sequential()
model.add(LSTM(128, input_shape=(train_X.shape[1], train_X.shape[2]), return_sequences=True))
model.add(Dropout(0.2))
model.add(LSTM(64, return_sequences=True))
model.add(Dropout(0.2))
model.add(LSTM(32))
model.add(Dense(n_out))
model.compile(loss='mean_squared_error', optimizer='adam')

# fit network
history = model.fit(train_X, train_y, epochs=250, batch_size=32, validation_split = .2, verbose=2, shuffle=True)

# summarize history for loss
plt.plot(history.history['loss'])
plt.plot(history.history['val_loss'])
plt.title('model loss')
plt.ylabel('loss')
plt.xlabel('epoch')
plt.legend(['train', 'test'], loc='upper right')
plt.show()
```

```python
# make a prediction
yPred = model.predict(train_X)
yPred.shape
```

```python
yhat = yPred.reshape((yPred.shape[0] * yPred.shape[1], 1))
train_XR = train_X.reshape((train_X.shape[0] * train_X.shape[1], 7))

inv_yhat = np.concatenate((yhat, train_XR[:, -6:]), axis=1)
inv_yhat = scaler.inverse_transform(inv_yhat)
inv_yhat = inv_yhat[:,0]

y = train_y.reshape((train_y.shape[0] * train_y.shape[1], 1))
inv_y = np.concatenate((y, train_XR[:, -6:]), axis=1)
inv_y = scaler.inverse_transform(inv_y)
inv_y = inv_y[:,0]
```

```python
# calculate RMSE
rmse = np.sqrt(mean_squared_error(inv_y, inv_yhat))
print('Test RMSE: %.3f' % rmse)
```
